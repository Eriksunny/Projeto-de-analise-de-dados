```python
import pandas as pd
import re
```

```python
df = pd.read_csv('melbourne_housing_data.csv')
```

```python
# Agrupa os dados pelo vendedor
# Soma todas as colunas num√©ricasr
# Ordena os vendedores pela soma total de quartos
# Mostra os 10 primeiros resultados
df.groupby(['SellerG']).sum().sort_values('Rooms', ascending=False).head(10)

df.groupby('SellerG').agg(
    total_vendas=('Price', 'sum'),
    total_quartos=('Rooms', 'sum'),
    quantidade_imoveis=('Price', 'count'),
    preco_medio=('Price', 'mean')
).sort_values('total_vendas', ascending=False).head(10)
```
