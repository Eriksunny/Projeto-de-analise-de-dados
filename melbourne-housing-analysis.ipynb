```python
import pandas as pd
import re
```

```python
df = pd.read_csv('melbourne_housing_data.csv')
```

```python
df.head(5)
```

```python
#Filtro 1

#Selecionar casas com:
#- 3 quartos
#- Tipo 'h'
#- Preço menor ou igual a 500000
#

filter1 = df.loc[
    (df['Rooms'] == 3) & (df['Type'] == 'h') & 
    (df['Price'] <= 500000)
    ]

filter1
```

```python
#Filtro 2 

#Selecionar o endereço
#- Endereço específico

filter2 = df.loc[
    df['Address'].str.contains('Turner St')
    ]

filter2
```

```python
#Filtro 3
#Selecionar casas com:
#-Numero da casa 59
#-preço abaixo de 500000

filter3 = df.loc[
    df['Address'].str.contains('^59', flags=re.I) & 
    (df ['Price'] <= 500000)
    ]

filter3
```
